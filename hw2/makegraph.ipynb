```python
import pandas as pd
import numpy as np

import seaborn as sns

import matplotlib.pyplot as plt
```

# firefox

```python
phy1 = pd.read_csv('./firefox/phy.txt',header=None)

virt1 = pd.read_csv('./firefox/virt.txt',header=None)

shared1 = pd.read_csv('./firefox/shared.txt',header=None)

text_data1= pd.read_csv('./firefox/td.txt', sep='\t',header=None)


plt.rcParams['ytick.right'] = plt.rcParams['ytick.labelright'] = True
plt.rcParams['ytick.left'] = plt.rcParams['ytick.labelleft'] = False


phy1[0].plot()
plt.grid(True)
plt.title('Physical')
plt.xticks(np.arange(0, len(phy1)+1,10),rotation=90)
plt.xlabel('time*2(seconds)')
plt.ylabel('resident set size(page)')
plt.savefig('./firefox/graph/firefox_physical.png', bbox_inches='tight',dpi=300)
plt.show()

shared1[0].plot()
plt.grid(True)
plt.title('Shared')
plt.xticks(np.arange(0, len(shared1)+1,10),rotation=90)
plt.xlabel('time*2(seconds)')
plt.ylabel('shared pages(page)')
plt.savefig('./firefox/graph/firefox_shared.png', bbox_inches='tight', dpi=300)
plt.show()

virt1[0].plot()
plt.grid(True)
plt.title('Virtual')
plt.xticks(np.arange(0, len(virt1)+1,10),rotation=90)
plt.xlabel('time*2(seconds)')
plt.ylabel('total program size(page)')
plt.savefig('./firefox/graph/firefox_virtual.png',bbox_inches='tight', dpi=300)
plt.show()

text_data1[0].plot()
plt.grid(True)
plt.title('text(code)')
plt.xticks(np.arange(0, len(text_data1[0])+1,10),rotation=90)
plt.xlabel('time*2(seconds)')
plt.ylabel('number of pages')
plt.savefig('./firefox/graph/firefox_text.png',bbox_inches='tight', dpi=300)
plt.show()

text_data1[1].plot()
plt.grid(True)
plt.title('data(data(heap)+stack)')
plt.xticks(np.arange(0, len(text_data1[1])+1,10),rotation=90)
plt.xlabel('time*2(seconds)')
plt.ylabel('number of pages')
plt.savefig('./firefox/graph/firefox_data.png',bbox_inches='tight', dpi=300)
plt.show()

b=virt1[0]-shared1[0]
b.plot()
plt.grid(True)
plt.title('Size-Shared')
plt.xticks(np.arange(0, len(b)+1,10),rotation=90)
plt.xlabel('time*2(seconds)')
plt.ylabel('number of pages')
plt.savefig('./firefox/graph/firefox_Size-Shared',bbox_inches='tight', dpi=300)
plt.show()
```
